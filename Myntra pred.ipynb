```python
sfc = sf[['Description','Individual_category']]
for i, rows in sf.iterrows():
    c = rows['Individual_category']
    u = []
    s = []
    p = word_tokenize(c)
    #s.append(p[k] for k in p)
    sfc['Description'][i] = p
    u = ' '.join(ps.stem(k)for k in p)
    sfc['Individual_category'][i] = u
```

```python
sfc = sfc.rename({'Description':'Toke'}, axis = 1)
```

```python
#del f
def recom(text,gender):
    index = []
    for i,rows in sf.iterrows():
        c = rows['Toke']
        p = len(c)
        te = text.lower()
        for j in range (0,p):
            if(c[j] == te):
                index.append(i)
        f = pd.DataFrame()
    for i in range (0,len(index)):
        f[i] = sf.loc[index[i],]
    f=f.transpose()
    q = (f[f['category_by_Gender'] == gender].sort_values(['Sell Price','Ratings','Reviews'], ascending=[True,False,False]))
    return q
        
            
```
